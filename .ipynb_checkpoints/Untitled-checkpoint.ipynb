```python
%matplotlib inline
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread('./lena.ppm', cv.IMREAD_GRAYSCALE)
edges = cv.Canny(img, 100, 600)
cv.imshow("name", img)

plt.subplot(121), plt.imshow(img, cmap='gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(edges, cmap='gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
import os
import glob

image_path = './captcha'
images = [
    cv.imread(filename, cv.IMREAD_GRAYSCALE)
    for filename in glob.glob(image_path + '/*.png')
]
images_num = len(images)
```

```python
plt.imshow(images[0], cmap='Greys')
images[5]
```

```python
sum_image = np.zeros((36, 120), dtype=np.uint8)
for i in range(len(images)):
    sum_image = sum_image + (images[i] / images_num)
plt.imshow(sum_image, cmap='Greys')
sum_image
```

```python
_, noise = cv.threshold(sum_image, 200, 255, cv.THRESH_BINARY_INV)
plt.imshow(noise, cmap='Greys')
plt.show()
```

```python
for i in range(images_num):
    _, images[i] = cv.threshold(images[i], 200, 255, cv.THRESH_BINARY_INV)
plt.imshow(images[0], cmap='Greys')

test_result = images[0] - noise
plt.imshow(test_result, cmap='Greys')
plt.show()
```

```python
resized_images = []
for i in range(images_num):
    resized_images.append(images[i] - noise)
    resized_images[i] = cv.resize(resized_images[i], None, fx=8, fy=8, interpolation=cv.INTER_NEAREST)
print(resized_images[0].shape)
plt.imshow(resized_images[0], cmap='Greys')
plt.show()
```

```python
import random

kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE, (13,13))
kernel2 = cv.getStructuringElement(cv.MORPH_ELLIPSE, (2,2))
final_images = []
for i in range(images_num):
    final_images.append(cv.erode(cv.dilate(resized_images[i], kernel), kernel2))
#     final_images.append(cv.dilate(resized_images[i], kernel))

plt.imshow(final_images[0], cmap='Greys')
plt.show()
```

```python
resized_images = []
for i in range(images_num):
    resized_images.append(cv.resize(final_images[i], None, fx=0.25, fy=0.25, interpolation=cv.INTER_NEAREST))
    
plt.imshow(resized_images[0], cmap='Greys')
plt.show()
print(resized_images[0].shape)
```
